```python
midterm_1=[2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
midterm_1.sort()
midterm_frequency={}
for number in midterm_1:
    if number in midterm_frequency:
        midterm_frequency[number] +=1
    else:
        midterm_frequency[number]=1
for key, value in midterm_frequency.items():
    print(key, ' : ', value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = midterm_frequency.keys()
y = midterm_frequency.values()
plt.bar(x,y)
plt.title('Frequency Distribution for Midterm Question 1')
plt.xlabel('Number')
plt.ylabel('Frequency')
```
